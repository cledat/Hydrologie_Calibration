```python
for ID in OBJECTID_list:
    #### Create regional settings files 
    
    setting_file =ini_setting_folder+ 'settings_CWatM_template-netcdf-regional.ini'
    new_setting_file = ini_setting_folder+ 'regional_settings_files/settings_CWatM_template-netcdf-regional'+str(int(ID))+'.ini'
    f = open(setting_file,"r")
    template_xml = f.read()
    f.close()
    ## Modification du fichier settings file 
    template_xml_new=template_xml
    name = '/{}_wmobb_sub_region.tif'.format(int(ID))
    template_xml_new = template_xml_new.replace('%mask_map', name)
    template_xml_new = template_xml_new.replace('%pfaf_id', str(int(ID)))
    template_xml_new = template_xml_new.replace('%out_path', output_path+'/'+str(int(ID)))
    # os.touch(settings_folder+name_set+str(int(ID)) + '.ini')
     ## Ajout des points de gauges
    lon_list_nan = obs.where(obs.OBJECTID == ID).gauged_lon.values
    lat_list_nan = obs.where(obs.OBJECTID == ID).gauged_lat.values
    lon_list = [str(x) for x in lon_list_nan if not np.isnan(x)]
    lat_list = [str(x) for x in lat_list_nan if not np.isnan(x)]
    gauges_str = ''
    for i in range(len(lon_list)):
        gauges_str= gauges_str+' ' + lon_list[i] +' ' + lat_list[i]
    template_xml_new = template_xml_new.replace('%gauges', gauges_str)
    ## Ajout Parametres calibration 
    noms_parametres =list(parametres_calibration.keys())
    for name in noms_parametres :
        template_xml_new = template_xml_new.replace('%'+name,str(parametres_calibration[name] ))

    
    f = open(new_setting_file, "w+")
    f.write(template_xml_new)
    f.close() 


 ########## Creer fichier sortie adapte + test en turilisant run cwatM slurm 
    mkdir_ID = 'mkdir ' + output_path+'/'+str(int(ID))
    moove  = 'cd ' + '/work/ch0636/g300128/Hydrologie_Calibration'
    moovement = subprocess.Popen(moove, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)

    commande = 'sbatch ' + ini_setting_folder + '../run_cwatm_slurm.sh ' + new_setting_file
    region_folder = subprocess.Popen(mkdir_ID, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    processus = subprocess.Popen(commande, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    # Récupérer la sortie standard et les erreurs éventuelles
    sortie, erreur = processus.communicate()
    print(commande)
    # Afficher la sortie
    if processus.returncode == 0:
        print("Répertoire de travail actuel:", sortie.decode())
    else:
        print("Erreur:", erreur.decode())
        ## run sbatch ##
```
