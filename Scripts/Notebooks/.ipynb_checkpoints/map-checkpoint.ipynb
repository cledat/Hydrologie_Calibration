# Extract and adapt data

Input : Ctif region map + sub region map 

Output : One map with  the sub basins for each points

```python
map_path = '/work/ch0636/g300124/xx_cwatm/area-maps/hybas_eu_lev03_finalselection_5arcmin.tif'
```

```python
# Remplacez 'chemin_vers_votre_fichier.shp' par le chemin réel vers votre fichier shapefile
chemin_fichier_shapefile = '/work/ch0636/g300128/Hydrologie_Calibration/data/hybas/hybas_eu_lev04_v1c.shp'

# Lire le fichier shapefile en utilisant geopandas
gdf = gpd.read_file(chemin_fichier_shapefile)
```

```python
rio_map = rioxarray.open_rasterio(map_path, chunks=True)
ds = rio_map.rio.reproject(ccrs.PlateCarree())
```

```python
ds
```

```python
list_of_station_IDs=[1,2,3]
calibrated_gauges = np.ones(len(list_of_station_IDs))
calibrated = {
    'data' : calibrated_gauges,
    'dims' : ['id'],  # 'dims' doit être une liste
    'coords' : {'id': {'dims': 'id', 'data': list_of_station_IDs}}  # 'coords' doit être un dictionnaire de dictionnaires
}
calibrated_array = xr.DataArray.from_dict(calibrated)
calibrated_array.name = 'calibrated'
```

```python
bands = 1  # Nombre de bandes
x_size = len(ds.x)  # Taille de la dimension x
y_size = len(ds.y) 
data = np.ones((bands, x_size, y_size))


for i in range(x_size) :
    for j in range(y_size) :
        point_to_check = Point(ds.x.values[i], ds.y.values[j])
        point_in_region = gdf.contains(point_to_check)
        # region_index = point_in_region.idxmax() if point_in_region.any() else None
        if point_in_region.any():
            region_index = point_in_region.idxmax()
            data[0,i,j] =gdf.iloc[region_index]['PFAF_ID']
# Créer le xarray DataArray avec les coordonnées et les dimensions
```

```python
import xarray as xr
import numpy as np

# Définir la taille du tableau pour chaque dimension
bands = 1  # Nombre de bandes
x_size = len(ds.x)  # Taille de la dimension x
y_size = len(ds.y)   # Taille de la dimension y

# Créer un DataArray avec des uns


xarr = xr.DataArray(data, dims=['band', 'x', 'y'],
                    coords={'band':ds.band,
                            'x': ds.x,
                            'y': ds.y.values,
                            'spatial_ref': 0})
ds.name = 'structure'
xarr.name = 'sub_basin'
# Afficher le xarray DataArray créé
map_array = xr.merge([ds,xarr])
```

```python
map_array.sub_basin.sel(x=0,y=47,method='nearest')
```

```python
# A_lon, A_lat, B_lon, B_lat = -12 , 34 , 41 , 72

choosen_projection = ccrs.PlateCarree()   
fig,ax = plt.subplots(figsize=(10,10),subplot_kw={'projection': choosen_projection})
# ax.set_extent([A_lon, B_lon, A_lat, B_lat], crs = choosen_projection)
map_array.structure.where((map_array.sub_basin==2340)).plot(add_colorbar=False,ax=ax,x= 'x', y='y')
ax.coastlines(linewidth=1, linestyle='-')
ax.gridlines(draw_labels=True, dms=True, x_inline=False, y_inline=False)
```

```python

OBJECTID_list = data_europe.OBJECTID.to_numpy().flatten().tolist()

PFAF_eu = list(set(OBJECTID_list))
for PFAF_ID in PFAF_eu:
    out_name = '/work/ch0636/g300128/Hydrologie_Calibration/data/Region_map_WMOBB/{}_wmobb_sub_region.tif'.format(int(PFAF_ID))
    map_array.structure.where(map_array.sub_basin ==int(PFAF_ID),drop=True).rio.to_raster(out_name, chunks=True)
```
